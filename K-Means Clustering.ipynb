# K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = make_blobs(n_samples=1000, n_features=2, centers=4, shuffle=True, random_state=500, cluster_std=2.1)
```

```python
plt.scatter(data[0][:,0],data[0][:,1], marker='+')
plt.show()
```

```python
means = KMeans(n_clusters=4)
y = means.fit_predict(data[0])
y
```

```python
clusters = means.cluster_centers_
clusters
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(data[0][y == 0,0], data[0][y == 0,1], color='green')
plt.scatter(data[0][y == 1,0], data[0][y == 1,1], color='red')
plt.scatter(data[0][y == 2,0], data[0][y == 2,1], color='blue')
plt.scatter(data[0][y == 3,0], data[0][y == 3,1], color='purple')
plt.scatter(clusters[:,0], clusters[:,1], color='black', marker='*', s=80)
plt.show()
```

# Evaluando el num. de cluster en el grafico, en busca del codo(Elbow)

Podemos encontrar la formacion de codo en los parametros 3 y 4.

```python
Ks = []
K_range = range(1,10)
for k in K_range:
    means = KMeans(n_clusters=k)
    means.fit(data[0])
    Ks.append(means.inertia_)
    
plt.xlabel('num_K')
plt.ylabel('Sum of squared error')
plt.plot(K_range, Ks)
```
